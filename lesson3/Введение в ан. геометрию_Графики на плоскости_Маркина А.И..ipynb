# Введение в аналитическую геометрию

```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
from scipy.optimize import fsolve
import warnings

warnings.filterwarnings('ignore')
```

```python
#Задание_1


def vector_length(coordinates_list):
    return round(np.sqrt(sum(list(map(lambda x: float(x) ** 2, coordinates_list)))), 2)


coordinates = (input('Введите коoрдинаты вектора через пробел: ')).split(' ')
print(f'Длина вектора составляет: {vector_length(coordinates)}')
```

```python
# Задание_3.1

x = []
x1 = []
y = []
y2 = []
r = 1000
for i in range(1001):
    x.append(i)
    x1.append(-i)
    y.append(np.sqrt(r ** 2 - i ** 2))
    y2.append(-np.sqrt(r ** 2 - i ** 2))
plt.title('График окружности')    
plt.plot(x, y, x, y2, x1, y, x1, y2, color='red')
plt.axis('scaled')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.grid()
plt.show()
```

```python
# Задание_3.2

x = []
x1 = []
y = []
y2 = []
a = 50
b = 30
for i in range(1000):
    x.append(i)
    x1.append(-i)
    y.append(np.sqrt((1 - (i ** 2) / (a ** 2)) * b ** 2))
    y2.append(-np.sqrt((1 - (i ** 2) / (a ** 2)) * b ** 2))
plt.title('График эллипса') 
plt.plot(x, y, x, y2, x1, y, x1, y2, color='green')
plt.axis('scaled')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.grid()
plt.show()
```

```python
# Задание_3.3

x = []
x1 = []
y = []
y2 = []
a = 500
b = 300
for i in range(1000):
    x1.append(-i)
    x.append(i)
    y.append(np.sqrt(b ** 2 + (b ** 2 * (i ** 2 / a ** 2))))
    y2.append(-np.sqrt(b ** 2 + (b ** 2 * (i ** 2 / a ** 2))))
plt.title('График гиперболы') 
plt.plot(x, y, x, y2, x1, y, x1, y2, color='violet')
plt.axis('scaled')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.grid()
plt.show()
```

```python
# Задание_5.1

fig = figure()
ax = Axes3D(fig)

x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)

z1 = x + 4*y - 50
z2 = x + 4*y + 20

ax.plot_wireframe(x, y, z1, colors='orange', label='x + 4y - z - 50 = 0')
ax.plot_wireframe(x, y, z2, colors='blue', linestyle='-.', linewidth=1, label='x + 4y - z + 20 = 0')
ax.scatter(0, 0, 0,'z',50,'red', label='(0; 0)')
plt.legend(frameon=False)
show()
```

```python
# Задание_5.2

fig = figure()
ax = Axes3D(fig)

x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)

z = np.sqrt(2**2 * (1 - (x**2 / 3**2) - (y**2 / 5**2)))
z1 = -np.sqrt(2**2 * (1 - (x**2 / 3**2) - (y**2 / 5**2)))

ax.plot_surface(x, y, z)
ax.plot_surface(x, y, z1)
ax.scatter(0, 0, 0)
show()

# ---------------------------------

fig1 = figure()
ax = Axes3D(fig1)

x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)

z = (x**2 / 20**2) - (y**2 / 25**2)

ax.plot_surface(x, y, z)
ax.scatter(0, 0, 0)
show()
```

```python
# Задание_3.1

def to_decart(r, alfa):
    x = round(r * np.cos(alfa), 2)
    y = round(r * np.sin(alfa), 2)
    return f'({x}; {y})'
r = 2
alfa = 45
print(f'Полярным координатам ({r}; {alfa}) соответствуют декартовы координаты {to_decart(r, alfa)}')
```

```python
# Задание_4.1

# exp(x) + x*(1 – y) = 1  
# y = x**2 – 1

x = np.linspace(-3, 5, 1001)
plt.plot(x, (np.exp(x) - 1 + x) / x)
plt.plot(x, x**2 - 1)
plt.ylim(-2,20)
plt.grid()
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.show()

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x*(1 - y) - 1)

x1, y1 = fsolve(equations, (-2, 2))
x2, y2 = fsolve(equations, (1, 4))
x3, y3 = fsolve(equations, (4, 10))
print(f'Первый корень: ({x1},{y1})')
print(f'Второй корень: ({x2},{y2})')
print(f'Третий корень: ({x3},{y3})')
print(f'Четвертый корень: (0, -1)')

# Так же есть еще один корень - (0, -1) (так как х есть в знаменателе, проверяем, есть ли решение при х=0)
```
